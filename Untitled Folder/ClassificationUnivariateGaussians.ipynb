```python
def plotGaussian(mu,sig):
    N = 1000;
    a = 4; # catch the most information of the gaussian
    # mu-a*sig < x < mu + a*sig
    x = np.linspace(mu-a*sig,mu+a*sig,N)
    
    pdf = (1/np.square(2*np.pi*sig**2))*np.exp(-((x-mu)**2)/(2*sig**2))
    
    plt.plot(x,pdf)
    plt.grid()
    plt.title("p(x) for Normal Distribution")
    plt.xlabel("x")
    plt.ylabel("p(x)")
    return 
```

```python
plotGaussian(1,1)
```

Q3

```python
def plotGaussians(mu,sig,pi):
    a = 4
    N = 1000
    grid_min_vec = mu-a*sig
    grid_max_vec = mu+a*sig
    grid_min = min(grid_min_vec)
    grid_max = max(grid_max_vec)
    
    x = np.linspace(grid_min,grid_max,N)
    pdf_0 = pi[0]*(1/np.square(2*np.pi*sig[0]**2))*np.exp(-((x-mu[0])**2)/(2*sig[0]**2))
    pdf_1 = pi[1]*(1/np.square(2*np.pi*sig[1]**2))*np.exp(-((x-mu[1])**2)/(2*sig[1]**2))
    pdf_2 = pdf_0 + pdf_1
    
    labels = ["N(3,1)", "N(0,1)", "0.5*N(3,1) + 0.5*N(0,1)"];
    plt.figure().set_figwidth(10)
    plt.plot(x,pdf_0,x,pdf_1,x,pdf_2)
    plt.grid()
    plt.legend(labels,loc=1)
    plt.title("Plot Gaussians")
    plt.xlabel("x")
    plt.ylabel("p(x)")
    
    return 
```

```python
plotGaussians(np.array([3,0]),np.array([1,1]),np.array([1/2,1/2]))
```

```python
def plotGaussians_x0(mu,sig,pi,x0):
    a = 4
    N = 1000
    grid_min_vec = mu-a*sig
    grid_max_vec = mu+a*sig
    grid_min = min(grid_min_vec)
    grid_max = max(grid_max_vec)
    
    x = np.linspace(grid_min,grid_max,N)
    pdf_0 = pi[0]*(1/np.square(2*np.pi*sig[0]**2))*np.exp(-((x-mu[0])**2)/(2*sig[0]**2))
    pdf_1 = pi[1]*(1/np.square(2*np.pi*sig[1]**2))*np.exp(-((x-mu[1])**2)/(2*sig[1]**2))
    pdf_2 = pdf_0 + pdf_1
    
    pdf_0_x0 = pi[0]*(1/np.square(2*np.pi*sig[0]**2))*np.exp(-((x0-mu[0])**2)/(2*sig[0]**2))
    labels = ["N(3,1)", "N(0,1)", "0.5*N(3,1) + 0.5*N(0,1)"];
    plt.figure().set_figwidth(10)
    plt.plot(x,pdf_0,x,pdf_1,x,pdf_2)
    plt.plot(x0,pdf_0_x0,marker='x')
    plt.text(x0 + .2, pdf_0_x0, "(x0 = %.2f,p(x0) = %.4f)"%(x0,pdf_0_x0))
    plt.grid()
    plt.legend(labels,loc=1)
    plt.title("Plot Gaussians")
    plt.xlabel("x")
    plt.ylabel("p(x)")
    
    return 
```

```python
x0 = 0.25
plotGaussians_x0(np.array([3,0]),np.array([1,1]),np.array([1/2,1/2]),x0)
```
